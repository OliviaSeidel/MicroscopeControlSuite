# Example remote controll Hamamatsu EMCCD camera


To run this you need Python 3.7.11, Hamamatsu DCAM API, and pip install hamamatsu.  The script below runs through a few exposure times in Normal CCD mode. EMMCD an be accessed by switching Normal to EMCCD in the CCDMode argument.

```python
from CamHam import CamHam
import pylab as plt
```

```python
camera=CamHam(CCDMode='Normal')
```

```python
for ei in [0.5, 1.5, 5, 10]:
    camera.exposure=ei
    pic=camera.Snap(1)[0]
    plt.imshow(pic)
    plt.title("exposure = " + str(ei) + " sec")
    plt.clim(3000,10000)
    plt.show()
```
